### Importar, corregir Datos y unir

```r
datos= read_csv('train.csv')
```

```r
datos %>% glimpse
```

```r
sum(is.na(datos))
```

```r
colSums(is.na(datos))
```

```r
sum(is.na(datos$year_built))
```

```r
head(datos, n = 5)
```

```r
datos %>% 
  group_by(building_class) %>%
    summarise(
      Cantidad = n()
    )
```

```r
print(n = 63, datos %>% 
  map_dfr(.f = function(x)tibble(NumValoresUnicos = length(unique(x)),tipoDato = typeof(x)), .id = "Var") %>%
    arrange(NumValoresUnicos))
```

```r
datos %$% unique(State_Factor)
datos %$% unique(building_class)
```

```r
datos %>%
  group_by(State_Factor) %>%
  summarise(Frec = n()) %>%
  arrange(-Frec)%>%
  mutate(
    Porcentaje = round(Frec/sum(Frec),4) * 100,
    PorcentajeAcum = round(cumsum(Frec)/sum(Frec),4)*100
  
  )
```

```r
datos %>%
  group_by(building_class) %>%
  summarise(Frec = n()) %>%
  arrange(-Frec)%>%
  mutate(
    Porcentaje = round(Frec/sum(Frec),4) * 100,
    PorcentajeAcum = round(cumsum(Frec)/sum(Frec),4)*100
  
  )
```
